```python
def gen_zip_to_politic():
  for i in range(len(cd_to_politic_df)):
    if len(cd_to_politic_df.index[i].split()) != 2:
      continue
    state_name, cd = cd_to_politic_df.index[i].split()
    try:
      cd = int(cd)
    except:
      print("can't parse:", cd)
      continue
    political_party = cd_to_politic_df.political[i][0]
    # print(state_name, cd)
    state_abbr = us_state_to_abbrev[state_name]

    idx_arr = np.where((zip_to_cd_df["state_abbr"] == state_abbr) & (zip_to_cd_df["cd"] == cd)) 
    idx_tup = idx_arr[0]
    # print(idx_tup)

    # print("pol:", political_party)

    for i in range(len(idx_tup)): # len(idx_arr)
      idx = idx_tup[i]
      zip_code = zip_to_cd_df.at[idx, "zip"]
      # print(zip_code)
      zip_to_politic_map[zip_code] = political_party
      zip_to_politic_dict["ZIP"].append(zip_code)
      zip_to_politic_dict["POL_PARTY"].append(political_party)
```

```python
zip_to_cd_df = pd.read_csv(f"{BASE_URL}/zip_to_cd.csv")
```

```python
swapped_us_state_to_abbrev = {}
for key in us_state_to_abbrev:
  val = us_state_to_abbrev[key]
  swapped_us_state_to_abbrev[val] = key
```

```python
zip_to_cd_df.tail()
```

```python
def get_state_for_zip(zipcode):
  temp = zip_to_cd_df[zip_to_cd_df["zip"] == zipcode]
  state = temp.iloc[0]["state_abbr"]
  return swapped_us_state_to_abbrev[state]
```

```python
mh_output_df = {"ZIPCODE": [], "MEDIAN_INCOME": [], "AT_RISK": []}
mental_health_df = mental_health_df[mental_health_df["Year"] == "2020"]
mental_health_df.head()
```

```python
mental_health_df.head()
```

```python
for row in tax_returns_median_df.iterrows():
  try:
    zipcode = row[1]["ZIPCODE"]
    med_income = row[1]["MEDIAN_INCOME"]
    state = get_state_for_zip(zipcode)
    temp = mental_health_df[mental_health_df["State"] == state]
    mh_at_risk = temp.iloc[0]["At Risk"]
    
    mh_output_df["ZIPCODE"].append(zipcode)
    mh_output_df["MEDIAN_INCOME"].append(med_income)
    mh_output_df["AT_RISK"].append(mh_at_risk)
  except:
    continue
```

```python
output = pd.DataFrame.from_dict(mh_output_df)
```

```python
output.to_csv(f"{BASE_URL}/mh_output.csv")
```
